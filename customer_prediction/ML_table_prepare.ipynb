### Add reordered column in order_product table

For each customer, if the product is ordered for the, mark it as 0, otherwise mark it as 1.

```````python
import pandas as pd

order_products_path = 'E:\Sebnewrepo\Data/order_products.csv'
order_products = pd.read_csv(order_products_path)

orders_path = 'E:\Sebnewrepo\Data/orders.csv'
orders = pd.read_csv(orders_path)
```````

User id is needed here to complement the granularity

```````python
joint = pd.merge(order_products, orders, on = 'order_id', how = 'left')
joint.head(10)
```````

Only need User_id here, make a new dataframe

```````python
new_df = pd.DataFrame({
    'user_id': joint['user_id'],
    'product_id': joint['product_id'],
    'order_number': joint['order_number'],
    'order_id': joint['order_id'], 
    'add_to_cart_order': joint['add_to_cart_order']
})
new_df.head(20)
```````

```````python
#easier than SQL partition query
new_df['rank'] = new_df.groupby(['user_id','product_id'])['order_number'].rank()
new_df.head(20)
```````

```````python
#new_df.sort_values(by = ['user_id','product_id','rank'], ascending = [False,False,False]).head(30)
def reordered(rank):
    if rank == 1:
        n = 0
    else:
        n = 1
    return n
new_df['reordered'] = new_df['rank'].apply(reordered)
new_df.head(20)
```````

```````python
#arrange a new data frame for order_product_new
output = pd.DataFrame({
    'order_id': new_df['order_id'],
    'product_id': new_df['product_id'],
    'add_to_cart_order': new_df['add_to_cart_order'],
    'reordered': new_df['reordered']
})
output.head(20)
```````

```````python
output['reordered'][output['reordered']==1]
```````

```````python
# output to the csv file
output.to_csv('E:\Sebnewrepo\Data\order_product_new.csv',index = False, header = True)
```````

### Add eval_set column in order table
This column is to divide the order into train, test, prior.
the order with train lable is the latest order for a user.
orders with prior lable are the previous order for a user, which are part of train set.
the order with test lable is the record needed to predict.

```````python
import numpy as np
# stats the distinct order number both in order_product and orders
m = order_products.order_id.unique()
print('distinct orders in order_products table:')
print(m.shape[0])
n = orders.order_id.unique()
print('distinct orders in orders table:')
print(n.shape[0])
```````

find the orders which is not exist in  order_products

```````python

dif = np.array(list(set(n).difference(set(m))))
print('the orders which is not exist in  order_products:')
print(dif)
print('How many orders:')
print(dif.shape[0])
```````

```````python
#make a joint table 
joint = pd.DataFrame({
    'order_id': dif,
    'count': np.ones(75000)
})
joint.head()
```````

```````python
orders.head(20)
```````

```````python
orders['ranking'] = orders.groupby(['user_id'])['order_number'].rank(ascending=False)
orders.head(20)
```````

```````python
joint1 = pd.merge(orders, joint, on = 'order_id', how = 'left')
joint1['count'] = joint1['count'].fillna(0)
joint1.head(100)
```````

```````python
# if rank = 0, then test, rank = 1, then train, rank >1, then prior.
joint1['ranking'] = joint1['ranking'] - joint1['count']
joint1.head(100)
```````

```````python
#this is a faster way
def eval_set(a):
    if a == 0:
       n = 'Test'
    elif a == 1.0:
       n = 'train' 
    else:
       n = 'prior'
    return n
joint1['eval_set'] = joint1['ranking'].apply(eval_set)
joint1.head(100)
```````

#### counting the result

```````python
print('Number of records with prior lable:')
print(joint1['eval_set'][joint1['eval_set'] == 'prior'].count())
print('Number of records with train lable:')
print(joint1['eval_set'][joint1['eval_set'] == 'train'].count())
print('Number of records with Test lable:')
print(joint1['eval_set'][joint1['eval_set'] == 'Test'].count())
```````
