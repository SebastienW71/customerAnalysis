This workbook is to apply the regex lab to clean the product.csv table

```python
import pandas as pd
import re
import string 

products_path = 'E:\Sebnewrepo\Data/products.csv'
products = pd.read_csv(products_path)
```

```python
products.head()
```

```python
products.shape
```

```python
# the replacement function
def clean(str):
    m = re.sub(r'\"','', str)
    m = re.sub(r'\\','', m)
    m = re.sub(r'\,','', m)
    return m
```

```python
# use pandas .apply() function to process records in batch
products['product_name'] = products.product_name.apply(clean)
```
