```python
from scipy.stats import levy_l
import numpy as np
import matplotlib.pyplot as plt
```

```python
def levy_left(a,b):
    num_l = levy_l.numargs
    rv = levy_l(a,b) 
    quantile = np.arange (0.03, 2, 0.21) 
    R = levy_l.rvs(a, b) 
    R = levy_l.pdf(a, b, quantile) 
    distribution = np.linspace(0, np.maximum(rv.dist.b, a))
    return distribution,rv

def summetry(x,array):
    a_bi = 2 * x - array
    return a_bi
```

```python
d,rv = levy_left(5,3)
print(d)
m = rv.pdf(d)
plot = plt.plot(d, rv.pdf(d)) 
```

```python
d1 = summetry(2.5,d)
print(d1)
plot = plt.plot(d1, rv.pdf(d)) 
```

```python
log_dx = np.log(d+1)
log_dy = np.log(rv.pdf(d)+1 )
plot = plt.plot(log_dx,m)
```
