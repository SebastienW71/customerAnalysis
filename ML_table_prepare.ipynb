### Add reordered column in order_product table

For each customer, if the product is ordered for the, mark it as 0, otherwise mark it as 1.

```python
import pandas as pd

order_products_path = 'E:\Sebnewrepo\Data/order_products.csv'
order_products = pd.read_csv(order_products_path)

orders_path = 'E:\Sebnewrepo\Data/orders.csv'
orders = pd.read_csv(orders_path)
```

User id is needed here to complement the granularity

```python
joint = pd.merge(order_products, orders, on = 'order_id', how = 'left')
joint.head(10)
```

Only need User_id here, make a new dataframe

```python
new_df = pd.DataFrame({
    'user_id': joint['user_id'],
    'product_id': joint['product_id'],
    'order_number': joint['order_number'],
    'order_id': joint['order_id'], 
    'add_to_cart_order': joint['add_to_cart_order']
})
new_df.head(20)
```

```python
#easier than SQL partition query
new_df['rank'] = new_df.groupby(['user_id','product_id'])['order_number'].rank()
new_df.head(20)
```

```python
#new_df.sort_values(by = ['user_id','product_id','rank'], ascending = [False,False,False]).head(30)
def reordered(rank):
    if rank == 1:
        n = 0
    else:
        n = 1
    return n
new_df['reordered'] = new_df['rank'].apply(reordered)
new_df.head(20)
```

```python
#arrange a new data frame for order_product_new
output = pd.DataFrame({
    'order_id': new_df['order_id'],
    'product_id': new_df['product_id'],
    'add_to_cart_order': new_df['add_to_cart_order'],
    'reordered': new_df['reordered']
})
output.head(20)
```

```python
# output to the csv file
output.to_csv('E:\Sebnewrepo\Data\order_product_new.csv',index = False, header = True)
```
